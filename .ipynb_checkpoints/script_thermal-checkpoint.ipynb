# Tutorial: Run SCRIPT with thermal evolution and feedback included

```python
from __future__ import print_function

import numpy as np
import pylab as plt
import os, glob
import configparser

import script
```

### The N-body snapshots must be present before this tutorial is run. For MUSIC snapshots, simply edit the file **'run_2lpt_music.py'** and run **'python3 run_2lpt_music.py'**.

```python
## directory where 2LPT outputs are stored
two_lpt_outpath = './MUSIC_outputs/N64_L64.0'
outroot = 'snap' ## root of the output snapshots

ngrid = 16 ## number of grid cells in each direction used for generating the maps

### Model parameters, see Maity & Choudhury (arXiv:2110.14231) for definitions.
### ionization efficiency zeta (z)
zeta_0 = 25.0
alpha = 1.4
z_0 = 5.5

### clumping factor
clumping = 3.0

### reionization temperature
T_reion = 1.7e4 ### K

### escape fraction fesc(M)
fesc_0 = 0.06
beta = -0.4
M_0 = 8.e10 ### Msun
```

```python
### Read data files, provided in the directory './data_files'. Needed to compare the model with observations

zend_min = 5.3 ### Lower limit on end of reionization

tau_e_filename = './data_files/tau_e.txt' ### CMB optical depth
dark_pixels_filename = './data_files/Dark_pixels.txt' ### Dark pixels
T0_gamma_filename = './data_files/T0_gamma.txt' ### T0 and gamma

zUVLF_arr = np.array([7.0, 6.0]) ### redshifts where UVLF data is used
UVLF_filename_list = [] ### list of data files
for i, zUVLF in enumerate(zUVLF_arr):
    zchar = '{:.1f}'.format(zUVLF)
    zchar = zchar.replace('.', 'p')
    UVLF_filename = './data_files/UV_lumfun_z' + zchar + '.txt'
    UVLF_filename_list.append(UVLF_filename)    
```

```python
### Class object to compute the evolution.
thermal_evolution = script.thermal_ion_evolution(two_lpt_outpath, ngrid, 
                                                 zend_min=zend_min, 
                                                 tau_e_filename=tau_e_filename, 
                                                 dark_pixels_filename=dark_pixels_filename, 
                                                 T0_gamma_filename=T0_gamma_filename, 
                                                 zUVLF_arr=zUVLF_arr, 
                                                 UVLF_filename_list=UVLF_filename_list)##, verbose=True)

thermal_evolution.data_root = 'snap'
thermal_evolution.outpath = two_lpt_outpath + '/script_files'  ## directory where the script-related files would be stored

thermal_evolution.zeta_0 = zeta_0
thermal_evolution.alpha = alpha
thermal_evolution.T_reion = T_reion
thermal_evolution.fesc_0 = fesc_0
thermal_evolution.beta = beta
thermal_evolution.clumping = clumping
thermal_evolution.z_0 = z_0
thermal_evolution.M_0 = M_0

thermal_evolution.feedback = 'step' ### other options are 'Jeans-gradual' and 'no'

### Computing thermal properties of the sub-grid density elements
thermal_evolution.compute_temp_element = True
thermal_evolution.element_nsamp = 6 ### number of elements per grid cell
thermal_evolution.element_seed = 12345
thermal_evolution.element_sigma = 1.0 ### the standard deviation of density elements at the initial redshift

### Redshifts for which the data will be written in a file, can be used later
thermal_evolution.zsave_arr = np.array([7, 6, 5])
thermal_evolution.save_root = 'test'


# thermal_evolution.M_UV_edges_list = []
# thermal_evolution.M_UV_edges_list.append(np.linspace(-8.0, -22.0, 141))
# thermal_evolution.M_UV_edges_list.append(np.linspace(-8.0, -22.0, 141))

#thermal_evolution.M_UV_min = -21.0
```

```python
%%time

### Compute the evolution
### May take longer when run first time because of the density, velocity and halo fields being computed
thermal_evolution.thermal_evolution()

### Also calculate the chi-square for the model given the data sets
thermal_evolution.all_chisq()

print ('chisq, tau_e = ', thermal_evolution.chisq, thermal_evolution.tau_arr[0])
```

```python
## Plot the evolution of physical quantities

fig = plt.figure(figsize=(15, 10))
axs = fig.subplots(2, 2)

ax_Q = axs[0,0]  ### reionization history QHII
ax_N = axs[0,1]  ### emissivity of ionizing photons
ax_T = axs[1,0]  ### T0
ax_g = axs[1,1]  ### gamma

ax_Q.plot(thermal_evolution.z_arr, thermal_evolution.QHII_mean_arr)
ax_T.plot(thermal_evolution.z_arr, thermal_evolution.T0_arr)
ax_g.plot(thermal_evolution.z_arr, thermal_evolution.gamma_arr)
ax_N.plot(thermal_evolution.z_arr[0:], np.log10(thermal_evolution.emissivity_mean_arr[0:] + 1.e-16) + 50)

ax_Q.set_ylabel(r'$Q_{\mathrm{HII}}$')
ax_Q.axvline(thermal_evolution.zend_min, c='r', ls=':') ### the end of reionization constraint
ax_Q.errorbar(thermal_evolution.z_DP_arr, 1 - thermal_evolution.xHI_UL_arr, yerr=thermal_evolution.sigma_xHI_arr, lolims=True, fmt='o', c='r') ### the dark pixel data
#ax_Q.legend()

ax_T.set_ylabel(r'$T_0$ (K)')
ax_T.errorbar(thermal_evolution.z_T0_arr, thermal_evolution.T0_obs_arr, yerr=thermal_evolution.sigma_T0_obs_arr, fmt='o', c='r') ### T0 data

ax_g.set_ylabel(r'$\gamma$')
ax_g.errorbar(thermal_evolution.z_T0_arr, thermal_evolution.gamma_obs_arr, yerr=thermal_evolution.sigma_gamma_obs_arr, fmt='o', c='r') ### gamma data

ax_N.set_ylabel(r'$\dot{n}_{\mathrm{ion}} (\mathrm{s}^{-1} \mathrm{cMpc}^{-3})$')

#plt.yscale('log')

for i in range(2):
    for j in range(2):
        axs[i,j].set_xlim(5.0, 14.0)
        axs[i,j].set_xlabel(r'$z$')
ax_N.set_ylim(49.5,51.5)

plt.show()
```

```python
### Plot the UV luminosity function at the redshifts where data is available

fig = plt.figure(figsize=(14, 6))
axs = fig.subplots(1, 2)

for i, zUVLF in enumerate(thermal_evolution.zUVLF_arr):
    #file_savez = outpath + '/' + save_root + '_z' + '{:06.2f}'.format(zUVLF) + '.npz'
    #all_arr = np.load(file_savez)

    loop = np.argmin(np.abs(zUVLF - thermal_evolution.z_arr))
    z = thermal_evolution.z_arr[loop]
    #print (i, loop, file_savez)

    #M_UV_bin = all_arr['M_UV_bin']
    #Phi_UV = np.mean(all_arr['Phi_UV'], axis=(0,1,2))

    #axs[i].plot(M_UV_bin, Phi_UV)
    axs[i].plot(thermal_evolution.M_UV_bin_list[i], thermal_evolution.Phi_UV_bin_list[i])
    axs[i].errorbar(thermal_evolution.M_UV_obs_list[i], thermal_evolution.phi_obs_list[i], yerr=(thermal_evolution.phi_obs_lo_list[i], thermal_evolution.phi_obs_hi_list[i]), fmt='s', c='r')

    axs[i].set_ylim(1.e-5,10.0)
    axs[i].set_yscale('log')

    axs[i].set_title(r'$z \approx ' + '{:.2f}'.format(zUVLF) + '$')
    
    axs[i].set_xlabel(r'$M_{\mathrm{UV}}$')
    axs[i].set_ylabel(r'$\Phi (\mathrm{cMpc}^{-1} \mathrm{mag}^{-1})$')
    #axs[i].legend()

fig.tight_layout()
plt.show()
```

```python
### Let us look at the temperature-density phase plot

fig = plt.figure(figsize=(14, 3.5))
axs = fig.subplots(1, 3)

### find the saved files
for i, zsave in enumerate(thermal_evolution.zsave_arr):
    loop = np.argmin(np.abs(zsave - thermal_evolution.z_arr))
    zsave = thermal_evolution.z_arr[loop]
    
    file_savez = thermal_evolution.outpath + '/' + thermal_evolution.save_root + '_z' + '{:06.2f}'.format(zsave) + '.npz'
    all_arr = np.load(file_savez)
    
    
    #print (loop, z_arr[loop])
    
    if (i == 0):
        element_nsamp = np.shape(all_arr['element_densitycontr'])[0]
        z_re_element_arr = np.tile(thermal_evolution.z_re_arr, [element_nsamp,1,1,1])
    
    #print (np.shape(np.ravel(all_arr['element_densitycontr'])), np.shape(np.ravel(all_arr['temp_density_element'][0,:,:,:,:])))
    #sc = axs[i].scatter(1 + np.ravel(all_arr['densitycontr']), np.ravel(all_arr['temp']), s=1.0, c=np.ravel(z_re_arr))
    
    mask = np.logical_and(all_arr['element_densitycontr'] >= -0.9, all_arr['element_densitycontr'] <= 9.)
    sc = axs[i].scatter(1 + np.ravel(all_arr['element_densitycontr'][mask]), np.ravel(all_arr['temp_density_element'][0,:,:,:,:][mask]), s=1.0, c=np.ravel(z_re_element_arr[mask]))
    axs[i].plot(1 + np.ravel(all_arr['element_densitycontr'][mask]), thermal_evolution.T0_arr[loop] * (1 + np.ravel(all_arr['element_densitycontr'][mask])) ** (thermal_evolution.gamma_arr[loop] - 1), c='red')#, marker='.')
    
    cbar = fig.colorbar(sc, ax=axs[i])
    cbar.set_label(r'$z_{\mathrm{re}}$')

    axs[i].set_xscale('log')
    axs[i].set_yscale('log')
    axs[i].set_xlim(0.1, 10.0)
    axs[i].set_ylim(0.1e4, 1.e5)

    axs[i].set_xlabel(r'$\Delta$')
    axs[i].set_ylabel(r'$T$ (K)')
    
    axs[i].set_title(r'$z = ' + '{:.2f}'.format(zsave) + '$')
    
    txt = r'$T_0 = ' + '{:.2f}'.format(thermal_evolution.T0_arr[loop]) + '$K, $\gamma = ' + '{:.2f}'.format(thermal_evolution.gamma_arr[loop]) + '$'
    axs[i].text(0.1,0.9, txt, transform=axs[i].transAxes)
    
fig.tight_layout()
plt.show()
```
